```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
from PIL import Image
import requests
from io import BytesIO
```

```python
# first make our sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

# define our figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))

# now call both bar and plot elements to the same axes (ax)
ax.bar(x, y1,
color = "purple")
ax.plot(x, y2,
color = "blue")
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "purple")

ax.errorbar(x, y2, 
        yerr = y2_sd,
        fmt = "none")
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "blue")

ax.errorbar(x, y2,
        yerr = y2_sd,
        fmt = "none",
        ecolor= "purple",
        elinewidth= 4,
        capsize = 6,
        capthick= 4 )
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "blue")

ax.errorbar(x, y2,
        yerr = y2_sd,
        fmt = "none",
        ecolor= "orange",
        elinewidth= 4,
        capsize = 6,
        capthick= 4, 
        errorevery= 2)
```

```python
response = requests.get('https://upload.wikimedia.org/wikipedia/commons/9/91/Cartoon.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2,
color = "green")
ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])
ax_image.imshow(image)
ax_image.axis("off")
plt.show()
```
