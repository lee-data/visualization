```python
# Import necessary libraries 
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt

import pandas as pd
import scipy
```

```python
# Set random seed for reproducibility 
np.random.seed(613)

# Generate numbers for x from 1 to 50 but not including 50. 
x = np.arange(50)

# Create 50 random numbers for y, ranging from value 0 to 100. 
y = np.random.randint(0,100,50)

# Syntax for the scatter plot. 
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y)
```

```python
# Bar plot. 
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
# Horizontal bar plot. 
fig, ax = plt.subplots(figsize=(5,3))
ax.barh(x,y)
```

```python
# Line plot
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
# Histogram
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y) 
```

```python
# Line plot
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

# Set labels and title
ax.set_title('Total growth over time')
ax.set_ylabel('Total Growth')
ax.set_xlabel('Years since start')
```

```python
# Customize the plot with fonts, colors, sizes. 
font1 = {'family':'sans-serif', 'color':'blue', 'size':20}
font2 = {'family':'monospace', 'color':'green', 'size':14}

# Line plot with predefined customized fonts
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

# Set labels and titles
ax.set_title('Total Growth Over Time', fontdict= font1)
ax.set_ylabel('Total growth', fontdict= font2, loc = 'top')
ax.set_xlabel('Years since start', fontdict= font2, loc = 'right')
```

```python
# Customize the plot with different markers and colors
fig, ax = plt.subplots(figsize=(5,3))

ax.scatter(x,y,
          marker = '+',
          color = 'indigo'
          )
```

```python
# Customize the plot with fonts, colors, sizes, line style, line width, etc.
fig, ax = plt.subplots(figsize=(5,3))

ax.plot(x,y,
          marker = '+',
          color = 'indigo',
          linestyle = '--',
          linewidth = 2, 
         markeredgecolor = 'orange',
         markerfacecolor = 'yellow'
          )

ax.grid()
```

```python
# Create figure and axes
fig, ax = plt.subplots(figsize=(5, 3))

# Create a line plot with custom markers and line style
ax.plot(x, y,
          marker='+',
          color='cyan',
          linestyle = ':',
          linewidth=2, 
          markeredgecolor='orange',
          markerfacecolor='yellow')  # Correctly close the function call

# Customize grid to show only horizontal lines with specific color and style
ax.grid(True, which='both', axis='y', color='gray', linestyle='-', linewidth=0.5)
```

```python
# Class exercise: replicate a visual from a public link with the given public code. 


import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})

# Create the data
rs = np.random.RandomState(1979)
x = rs.randn(500)
g = np.tile(list("ABCDEFGHIJ"), 50)
df = pd.DataFrame(dict(x=x, g=g))
m = df.g.map(ord)
df["x"] += m

# Initialize the FacetGrid object
pal = sns.cubehelix_palette(10, rot=-.25, light=.7)
g = sns.FacetGrid(df, row="g", hue="g", aspect=15, height=.5, palette=pal)

# Draw the densities in a few steps
g.map(sns.kdeplot, "x",
      bw_adjust=.5, clip_on=False,
      fill=True, alpha=1, linewidth=1.5)
g.map(sns.kdeplot, "x", clip_on=False, color="w", lw=2, bw_adjust=.5)

# passing color=None to refline() uses the hue mapping
g.refline(y=0, linewidth=2, linestyle="-", color=None, clip_on=False)


# Define and use a simple function to label the plot in axes coordinates
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, fontweight="bold", color=color,
            ha="left", va="center", transform=ax.transAxes)


g.map(label, "x")

# Set the subplots to overlap
g.figure.subplots_adjust(hspace=-.25)

# Remove axes details that don't play well with overlap
g.set_titles("")
g.set(yticks=[], ylabel="")
g.despine(bottom=True, left=True)
```
