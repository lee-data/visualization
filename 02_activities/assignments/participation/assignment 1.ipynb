```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import seaborn
```

```python
# Set seed for reproduciblity
np.random.seed(613)

# Generate integers from 1to 49 on the x axis
x = np.arange(50)

# Generate 50 random data on the y axis, ranging from 0 to 99.
y = np.random.randint(0,100,50)

# Plot the data on a scatter plot 
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y)
```

```python
# Bar plot
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
# Line plot
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
# Histogram
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
# Line plot 
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

# Set labels and title
ax.set_title("Total growth over time")
ax.set_xlabel("Total Growth")
ax.set_ylabel("Years since start")
```

```python
# Customize fonts, colors, sizes
font1 = {"family": "sans-serif","color":"blue", "size":20}
font2 = {"family": "monospace","color":"green", "size":14}

# Line plot with predefined customized fonts
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title("Total growth over time", fontdict = font1)
ax.set_ylabel("Total growth", fontdict = font2)
ax.set_xlabel("Years since start", fontdict = font2)
```

```python
# Class Coding Activity 
# Requirement: Copy the provided code and attempt to replicate the output in Python

# Import necesesary libraries
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

# Set data: 
df=pd.DataFrame({                                # Initialize a new df 
    'x_values': range(1,101),                    # Create a series of numbers from 1 to 100 on the x-axis
    'y_values': np.random.randn(100)             # Generate an array of 100 random numbers from a standard normal distribution (mean = 0, standard deviation = 1).
                * 15                             # Scale these random numbers by 15.
                + range(1,101),                  # Add the series of numbers from 1 to 100 to the scaled random numbers. This means each value in the series will be added to a corresponding random number, resulting in a new array of 100 values.
    'z_values': (np.random.randn(100)            # (simimilar to what we did for y-values)
                 * 15
                 + range(1,101))*2 })
 

# Create a subplot grid with 1 row and 2 columns (ie. "12"). 
# The 1st subplot in the grid (ie. the left subplot): the graph of x-values vs y-values: 
plt.subplot(121)
plt.plot( 'x_values', 'y_values', data=df, marker='o', alpha=0.4)                                        # "alpha" is the transparency level (ranging from 0 to 1, where 0 is fully transparent and 1 is fully opaque)
plt.title("A subplot with 2 lines")
 
# The 2nd subplot (ie. the right subplot)
plt.subplot(122)
plt.plot( 'x_values','z_values', data=df, linestyle='none', marker='o', color="orange", alpha=0.3)      # "alpha" is the transparency level (ranging from 0 to 1, where 0 is fully transparent and 1 is fully opaque)
 
# Plot
plt.show()


# Question: Recall the aesthetic, substantive, and perceptual qualities of data visualizations. Does your visualization of choice succeed in each area?

# My answer below: 

# Aesthetic: 
    # Using different colors for different subplots (blue and orange) helps make it easy to distinguish between the 2 datasets.
    # The markers (marker='o') help viewers to see distinct data points better.

# Substantive:
    # The visualization shows 2 data distributions: x-values vs y-values on the left subplot; and x-values vs z-values on the right subplot.

# Perceptual:
    # The perceptual clarity is strong with distinct colors and markers.  
```

```python
# Customize marker and color: 
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y,
          marker ="+", # "o"/ "+"
          color = "cyan")
```

```python
# Line style
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
          marker ="+", # "o"/ "+"
          color = "cyan",
           linestyle = "--",
           linewidth = 2)
```

```python
# Explore different marker, color, linestyle, linewidth, marker's edge color, and marker's face color
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
          marker ="*", 
          color = "cyan", # line color
          linestyle = "--",
          linewidth = 2,
           markeredgecolor = "yellow", # marker edge color
           markerfacecolor = "orange") # marker inside color
```

```python
# Explore adding grid lines
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
          marker ="*", 
          color = "cyan", # line color
          linestyle = "--",
          linewidth = 2,
           markeredgecolor = "yellow", # marker edge color
           markerfacecolor = "orange") # marker inside color

# Add grid 
ax.grid()           # both vertical and horizontal grid lines

#ax.grid(axis = "x") # vertical grid lines
#ax.grid(axis = "y") # horizontal grid lines
```
